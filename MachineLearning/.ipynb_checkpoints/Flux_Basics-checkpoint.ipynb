```julia
d1 = MvNormal([1 0.5
              0.5 1])
d2 = MvNormal([2,4],[1 0.5;0.5 1])
# x1 = rand(d1,1000)
# scatter(x1'[:,1],x1'[:,2])
```

```julia
# Contour Plot - True posterior
gθ1      = -10:0.1:10
gθ2      = -10:0.1:10
gridθ1   = repeat(reshape(gθ1, 1, :), length(gθ2), 1)
gridθ2   = repeat(gθ2, 1, length(gθ1))
mamvn1(x,y) = pdf(d1,[x,y])
mamvn2(x,y) = pdf(d2,[x,y])
z = map(mamvn1, gridθ1, gridθ2)
w = map(mamvn2, gridθ1, gridθ2)
p1 = contour(gθ1,gθ2, mamvn1, fill=false,c=:black,levels=8,
    xlabel=L"\theta_1", ylabel=L"\theta_2",colorbar=false)
contour!(gθ1,gθ2, mamvn2, fill=false,c=:black,levels=8,
    xlabel=L"\theta_1", ylabel=L"\theta_2",colorbar=false)
```

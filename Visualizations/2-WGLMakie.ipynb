# Makie para Web

Vamos mostrar como utilizar o WGLMakie, o backend voltado para visualizações na web. Por ser para a web,
seu uso não é tão direto quanto o CairoMakie e o GLMakie.

```julia
using Pkg
Pkg.activate(".")
using JSServe
using WGLMakie
```

```julia
Page(exportable=true, offline=true);
```

```julia
f = scatter(rand(10,2))
```

```julia
N = 60
function xy_data(x, y)
    r = sqrt(x^2 + y^2)
    r == 0.0 ? 1f0 : (sin(r)/r)
end
l = range(-10, stop = 10, length = N)
z = Float32[xy_data(x, y) for x in l, y in l]
p = surface(
    -1..1, -1..1, z,
    colormap = :Spectral
)
```

```julia
grid = RectilinearGrid(0:10:100,0:10:100);
# grid = RectilinearGrid(0:0.1:1,0:0.1:1);
```

```julia
svg = m("svg", width=500, height=500);

function plotline(points,svg)
    for i in 1:length(points)
        p1 = points[i]
        if i == length(points)
            p2 = points[1]
        else
            p2 = points[i+1]
        end
        
        h = Hyperscript.attrs(svg)["height"]
        w = Hyperscript.attrs(svg)["width"]
        svg = svg(m("line", x1=p1[1], y1=p1[2], x2=p2[1], y2=p2[2], stroke="black"))
        svg = svg(transform = "scale(1 -1) translate($(w/2) $(h/2))")
    end
    return svg
end

for g in grid
    svg = plotline(coordinates.(vertices(g)),svg)
end
svg
```

```julia
F(v) = [(exp(v[1])-1)*cos(2π*v[2]),(exp(v[1])-1)*sin(2π*v[2])]
```

```julia
F.(coordinates.(vertices(grid[100]))./100).*10
```

```julia
svg = m("svg", width=500, height=500);
for g in grid
    svg = plotline(F.(coordinates.(vertices(g))./100).*100,svg)
end
svg
```

```julia
function plotspline(points,svg)
    for i in 1:length(points)
        p1 = points[i]
        if i == length(points)
            p2 = points[1]
        else
            p2 = points[i+1]
        end
        
        h = Hyperscript.attrs(svg)["height"]
        w = Hyperscript.attrs(svg)["width"]
        svg = svg(m("line", x1=p1[1], y1=p1[2], x2=p2[1], y2=p2[2], stroke="black"))
        svg = svg(transform = "scale(1 -1) translate($(w/2) $(h/2))")
    end
    return svg
end
```

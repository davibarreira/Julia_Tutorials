```julia
n = cl.e3
K = circle(no, n,1)

pc = plotcircle(K, name="Circle")
R  = exp(cdual(K))
applyrotor(R,x) = R * x * inv(R)

C = circle(point(1.5,0,0),cl.e2,1)
Cr = applyrotor(R,C)
pc1 = plotcircle(C)
pcr = plotcircle(Cr)
plot([pc, pc1,pcr])
```
